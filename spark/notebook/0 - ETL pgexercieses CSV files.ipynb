# What's ETL or ELT?

ETL stands for Extract, Transform, Load. In the context of Spark, ETL refers to the process of extracting data from various sources, transforming it into a desired format or structure, and loading it into a target system, such as a data warehouse or a data lake.

Here's a breakdown of each step in the ETL process:

## Extract
This step involves extracting data from multiple sources, such as databases, files (CSV, JSON, Parquet), APIs, or streaming data sources. Spark provides connectors and APIs to read data from a wide range of sources, allowing you to extract data in parallel and efficiently handle large datasets.

## Transform
In the transform step, the extracted data is processed and transformed according to specific business logic or requirements. This may involve cleaning the data, applying calculations or aggregations, performing data enrichment, filtering, joining datasets, or any other data manipulation operations. Spark provides a powerful set of transformation functions and SQL capabilities to perform these operations efficiently in a distributed and scalable manner.

## Load
Once the data has been transformed, it is loaded into a target system, such as a data warehouse, a data lake, or another storage system. Spark allows you to write the transformed data to various output formats and storage systems, including databases, distributed file systems (like Hadoop Distributed File System or Amazon S3), or columnar formats like Delta Lake or Apache Parquet. The data can be partitioned, sorted, or structured to optimize querying and analysis.

Spark's distributed computing capabilities, scalability, and rich ecosystem of libraries make it a popular choice for ETL workflows. It can handle large-scale data processing, perform complex transformations, and efficiently load data into different target systems.

By leveraging Spark for ETL, organizations can extract data from diverse sources, apply transformations to ensure data quality and consistency, and load the transformed data into a central repository for further analysis, reporting, or machine learning tasks.

# Interview Questions

While completing the rest of the practice, try to answer the following questions:

## Concepts
- What is ETL? (Hint: Explain each step)

    ETL is the process of Extracting data from a source, Transforming it into the desired format, and Loading it into a destination system.
    - Extract — read raw data from a source (CSV, database, API, etc.)
    - Transform — clean, filter, join, reshape the data into the desired format
    - Load — write the transformed data to a destination (data warehouse, Delta table, etc.)

## Databricks
- What is Databricks?

    Databricks is a cloud-based data platform built on top of Apache Spark. It provides a managed environment for data engineering, data science, and machine learning — with notebooks, clusters, and storage all integrated in one workspace.

- What is a Notebook?

    A notebook is an interactive document that combines code (Python, SQL, Scala), text (Markdown), and output (tables, charts) in a single interface. In Databricks, notebooks run on a Spark cluster.

- What is DBFS?

    DBFS (Databricks File System) is a distributed file system built into Databricks. It abstracts cloud storage (like Azure Blob Storage) and makes files accessible via paths like /FileStore/tables/bookings.csv. It's the equivalent of HDFS on a Hadoop cluster.

- What is a cluster? 

    A cluster is a set of virtual machines (nodes) that Databricks provisions to run Spark jobs. It consists of a driver node (coordinates) and worker nodes (execute tasks). In Serverless mode, Databricks manages the cluster automatically.

- Is Databricks a data lake or a data warehouse?

    Databricks is a Lakehouse — it combines the flexibility of a data lake (stores raw files in any format) with the structure and performance of a data warehouse (Delta tables with ACID transactions, schema enforcement). It supports both.

## Managed Table
- What is a managed table in Databricks?

    A managed table is a table where Databricks controls both the metadata (schema, table name) AND the data storage. When you drop a managed table, both the metadata and the underlying data files are deleted.

- Can you explain how to create a managed table in Databricks?

    df.write.format("delta").saveAsTable("table_name")
    Or via SQL: CREATE TABLE table_name AS SELECT ...

- Can you compare a managed table with an RDBMS table? (Hint: Schema on read vs schema on write)

    RDBMS (schema on write) — you define the schema before inserting data. The database enforces types and constraints at write time.  
    Managed table in Databricks (schema on read) — you can load raw data first and apply the schema when reading it. More flexible but less strict.

- What is the Hive metastore and how does it relate to managed tables in Databricks?

    The Hive metastore is a catalog that stores metadata about tables — their names, column names, types, and where their data files are located. Databricks uses it (or Unity Catalog) so Spark knows how to find and interpret tables when you run spark.sql("SELECT * FROM members").

- How does a managed table differ from an unmanaged (external) table in Databricks? (Hint: Consider what happens to the data when the table is deleted)

    Managed table — Databricks manages both metadata and data. Drop the table → data is deleted.  
    External table — Databricks manages only the metadata. The data lives at a location you specify. Drop the table → only metadata is deleted, data files remain.

- How can you define a schema for a managed table?

    ```
    from pyspark.sql.types import StructType, StructField, StringType, IntegerType

    schema = StructType([
        StructField("memid", IntegerType(), True),
        StructField("surname", StringType(), True)
    ])  
    df = spark.read.schema(schema).csv("/FileStore/tables/members.csv")
    ```

## Spark
`df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(file_location)`
- What does the option("inferSchema", "true") do? 

    Spark automatically detects the data type of each column by scanning the data. Without it, all columns default to StringType.

- What does the option("header", "true") do?

    Spark treats the first row of the CSV as column names instead of data.

- How can you write data to a managed table?

    df.write.format("delta").mode("overwrite").saveAsTable("members")

- How can you read data from a managed table into a DataFrame?

    df = spark.table("members")
    # or
    df = spark.sql("SELECT * FROM members")
